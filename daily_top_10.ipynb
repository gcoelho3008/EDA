```python
import pandas as pd
df = pd.read_csv('dados/netflix_daily_top_10.csv')
df.head()
```

- Os dados são identificados da seguinte forma
Datas(Date time): As of e Netflix Realease Date
- Numéricos Inteiros(int64): Rank, Year to Date Rank, Days In Top 10 e Viewership Score
- Categóricos/Texto(Object/String): Title, Type(Movie/TV Show) e Netflix Exclusive
- Especiais: Last Week Rank contem hiféns, exige limpeza para tornar numérica


```python
# Periodo de análise feita
# Converter os dados para o formato de data
df['As of'] = pd.to_datetime(df['As of'])

# início, fim, duração
data_inicio = df['As of'].min()
data_fim = df['As of'].max()
duracao = data_fim - data_inicio

print(f"Período de análise: de {data_inicio.date()} a {data_fim.date()}")
print(f"Total de dias cobertos a base: {duracao.days} dias")
```

```python
# tamanho da base de dados
tamanho = df.shape
print(f"Tamanho da base de dados: {tamanho[0]} linhas e {tamanho[1]} colunas")
```

```python
# Verificar os dados nulos
df.isnull().sum()
df.info()
(df.isnull().sum() / len(df)) * 100
```

- 9 colunas estão 100% preenchidas (0.00% de nulos).

- Netflix Exclusive: Possui 35,22% de dados faltantes (apenas 4.599 linhas preenchidas das 7.100 totais).


```python
# Analisar os outliers
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))

# viewership Score
plt.subplot(1, 2, 1)
sns.boxplot(x=df['Rank'], color='blue')
plt.title('Rank de Audiencia')

# Day in top 10
plt.subplot(1, 2, 2)
sns.boxplot(x=df['Days In Top 10'], color='green')
plt.title('Dias no Top 10')

plt.tight_layout()
plt.show()

```

```python
# 1. Calculando os limites
Q1 = df['Viewership Score'].quantile(0.25)
Q3 = df['Viewership Score'].quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# 2. Filtrando a base
outliers_df = df[(df['Viewership Score'] < limite_inferior) | (df['Viewership Score'] > limite_superior)]

# 3. Exibindo os resultados
print(f"Limite Superior para ser Outlier: {limite_superior}")
print(f"Quantidade de Outliers: {len(outliers_df)}")

# Ver os 10 maiores outliers (os maiores sucessos)
outliers_df.sort_values(by='Viewership Score', ascending=False).head(10)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Days In Top 10', y='Viewership Score', hue='Type', alpha=0.6)
plt.title('Relação entre Dias no Top 10 e Pontuação')
plt.show()
```
